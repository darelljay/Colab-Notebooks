```python
output_a = "{:d}".format(52)

#특정 칸에 출력하기
output_b = "{:5d}".format(52)
output_c = "{:10d}".format(52)

#  빈칸을 0으로 채우가
output_d = "{:05d}".format(52)
output_e = "{:05d}".format(-52)

print("#기본")
print(output_a)
print("# 특정칸에 출력하기")
print(output_b)
print(output_c)
print("# 빈칸을 0으로 채우기")
print(output_d)
print(output_e)
```

```python
a = float(input("밑변의 길이를 입력하세요"))
b = float(input("높이의 길이를 입력해주세요"))
c2 = a**2+b**2
f = 2
while f<c2:
  if c2%f ==0:
    c2/=f
  else:
    f+=1
print(c2)
```
