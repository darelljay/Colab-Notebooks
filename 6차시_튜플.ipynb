**튜플의 연산자**

```python
# 튜플에서의 더하기(+)연산자 (연결)
tpl1 = (1, 2, 3, 4, 5)
tpl2 = (6, 7, 8, 9, 10)
print(tpl1+tpl2)
```

```python
# 튜플과 리스트를 더하면 (연결) ==> 튜플로 만들기
lst = [1, 2, 3, 4, 5]
tpl = (6, 7, 8, 9, 10)
tuple(lst)+tpl
```
