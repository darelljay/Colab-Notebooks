(예제)나이가 7세 미만일 경우에만 입장료 무료 메시지 출력하기

```python
age = int(input('나이를 입력하세요'))

if(age<7):
  print('무료입장입니다')
else:
  print('돈내세요!')  
```

###5) 필기 점수와 실기 점수를 입력 받아 두 점수가 모두 80점 이상인 경우에 '합격입니다'를 출력하는 프로그램을 작성하시오. 단 불합격인 조건에서는 아무것도 출력하지 않음.

```python
test1 = int(input('필기 점수를 입력하세요: '))
test2 = int(input('실기 점수를 입력하세요: '))
result = test1 + test2

if(result>=80):
  print('합격')
  
```
