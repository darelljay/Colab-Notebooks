# 함수  
파이썬의 함수는

1.일련의 코드를 담아 여러번 재사용할 수 있다.

2.인수를 매개변수로 받아 코드를 실행한 후 결과를 반환한다.  


> def 함수이름(매개변수):  
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행할 문장1  
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행할 문장2  
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return 결과값

함수 호출의 결과로 반환하는 값, 반환값(return)

```python
#매개 변수가 여러개인 경우 (매개 변수 순서를 지켜야 함)
#시작수 끝수를 입력 받아 3의 배수 합계를 구하시오. 
def sum_besu(start,end):
  sum = 0
  for i in range(start,end+1):
    if i%3==0:
     sum+=i
  print("3의 배수의 합: ",sum)
start = int(input("시작수: "))
end = int(input("끝수: "))
sum_besu(start,end)
```

```python
#키워드 인자를 사용하는 함수
def get_minus(x,y,z):
  minus = x-y-z
  return minus
print(get_minus(5,10,15))
print(get_minus(5,z=10,y=15))
```

```python
#기본 인자 값을 갖는 함수
def hello(msg ="안녕하세요"):
    print(msg + "!")

hello("오랜만이에요")
hello("이영희") 
hello()
```

```python
#일부 변수만 기본 인자 값을 갖는 함수
def hello(name,msg = "안녕하세요",feeling="♡"):
  print(name+ "님,"+ msg + feeling)

hello("아무개")
hello("김철수","오랜만이에요","♬")
```

```python
#두 수 중 큰 수 구하기
#함수 정의
def find_max(a,b):
  if(a>b):
    max=a
  else:
    max=b
    return max


#함수 호출
find_max(10,20)
```

```python
#짝수인지 판별하는 even 함수 만들기,bool 형으로 결과값 리턴
#함수 선언
def even(num):
  if num%2==0:
    result = True
  else:
    result = False
  return result
#함수 호출
print(even(17))
```

```python
#여러개의 값을 리턴하는 함수
def calculate_avg(a,b,c):
  sum=a+b+c
  return sum/3
avg = calculate_avg(1,5,9)
print("result of function call =",avg)
```
