**세트에서 특정 원소 제거하기**

```python
set1 = {1,2,3,4,5}
```

```python
# 만약 해당 원소값이 set내에 존재하지 않을 경우
```

```python
A = {1,2,3,5}
B = {3,5,6,8,9}
```

```python
# A와 B의 공통원소를 반환하는 교집합  (& ,intersection )
s3 = {3,6,9,12}
s4 = {4,8,12,16}
s3 & s4
s3.intersection(s4)
```

```python
# A와 B의 모든원소를 반환하는 합집합 ( |, union )
s3 | s4
s3.union(s4)
```

```python
# A집합에서 A와 B의 교집합을 제외한 원소를 반환하는 차집합 ( -, difference)
s3 - s4
s3.difference(s4)
```
