# **[도형넓이]**

**[삼각형의 넓이]**

키보드로 밑변의 길이와 높이를 입력 받아 삼각형의 넓이 구하기

image.png

삼각형의 넓이 = 밑변의 길이 * 높이 /2

예) area = width * height /2 

실행결과

삼각형의 밑변의 길이를 입력하세요: 40

삼각형의 높이를 입력하세요: 20

삼각형의 넓이 : 400.0 cm2

```python
width = float(input("삼각형의 밑변의 길이를 입력하세요: "))
height = float(input("삼각형의 높이를 입력하세요: "))
print((width*height)/2)
```

**[사다리꼴의 넓이 구하기]**

image.png

사다리꼴의 넓이 = ((밑변의 길이 + 윗변의 길이) * 높이)/2

예) area = ((top+bottom)*heght)/2

실행결과

사다리꼴의 윗변의 길이를 입력하세요: 55

사다리꼴의 밑변의 길이를 입력하세요: 22

사다리꼴의 높이를 입력하세요: 5

사다리꼴의 넓이 : 192.5 cm2

```python
top = int(input('사다리꼴의 윗변의 길이를 입력하세요: '))
bottom = int(input('사디리꼴의 밑변의 길이를 입력하세요: '))
height = int(input('사다리꼴의 넓이를 입력하세요: '))
print("사다리꼴의 넓이:",((top+bottom)*height)/2)
```

[킬로그램을 파운드로 변환]

[화씨를 섭씨로 변환]

섭씨 = 5/9 * (화씨 - 32)

예) C = 5/9 * (F-32)

실행결과

화씨 온도를 입력하세요: 33

---------------

화씨     섭씨

---------------

33.0     0.6

```python
h = float(input('화씨 온도를 입력하세요: '))
print('화씨:',h)
print('섭씨:',5/9*(h-32))
```

# **[거스름돈 계산하기]**

거스름돈 = 지불금액 - (물건값*구매개수)

예)change = pay - (price*num)

실행결과

지불 금액을 입력하세요: 5000

물건 값을 입력하세요: 200

구매 개수를 입력하세요: 3


지불 금액: 5000

물건 값: 200

구매 개수: 3

---------------

거스름돈은 4400원 입니다.


```python
cash = int(input('지불금액을 입력하세요: '))
price = int(input('물건값을 입력하세요: '))
amount = int(input('물건 갯수를 입력하세요: '))

print('지불금액',cash)
print('물건값: ',price)
print('물건 개수t: ',amount)
print('-'*100)
change = cash -(price*amount)
print('거스름돈은 ',change)
```
