#  반복문(while)



조건에 따라서 블록을 반복 실행하는 제어문이다.

while 조건: # 블록 실행 조건  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행할 문장1  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행할 문장2  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행할 문장3  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;...

while 다음에 제시되는 조건이 참인 동안에 수행할 문장을 반복해서 실행하게 됩니다.  
조건에는 앞서 조건문 if에서 논리형 자료 외에도 숫자형, 문자열형, 리스트형 등의 자료의 참과 거짓을 판단하는 방법이 동일하게 적용됩니다.


```python
#[예제] 아이디, 패스워드를 입력 받아 일치하면 인증 성공, 하나라도 틀리면 인증 실패를 출력 하시오.

id = "python"
pw = "1234"

print("아이디를입력하세요") 
a=input()
print("패스워드를 입력하세요")
b=input()

if( a== id and b==pw ):
  print("인증성공")
else:
  print("인증실패")

```

**반복문(while)의 다양한 형태**

while조건: 에서 조건에 올 수 있는 다양한 형태
반복문에서의 무한루프 와 해결

**반복문에서의 다양한 조건**

```python
# 리스트일때
a = [1,2,3,4,5]
while a:
  print(a.pop(),a)
```

```python
# 정수형일때
a = 5
while a:
  print(a)
  a-=1
  print(a)
  print('---' )
```

```python
#입력 값이  1인 동안 '맞습니다'를 출력하고 1이 아닌 값이 입력되면 종료
a = int(input("입력"))

while a==1:
  print('맞습니다.')
  a = int(input("입력"))
```

```python
# 입력 값이 숫자이면 '맞습니다'를 출력하고 아니면 종료
a = input("입력: ")

while a.isdigit( ):
  print('맞습니다')
  a = input('입력')
```
