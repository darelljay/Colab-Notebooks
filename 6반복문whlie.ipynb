###5)키보드로 파운드를 입력 받아 킬로그램으로 환산하는 프로그램을 작성하시오.

조건:

키보드로 -1을 입력할 경우에 프로그램을 종료함.

힌트:

킬로그램 = 파운드 x 0.453592

실행결과:

파운드를 입력하세요(종료:-1) : 554

554 파운드(lb)는 251.29 킬로그램(kg)입니다.

파운드를 입력하세요(종료:-1) : -1

프로그램이 종료되었습니다!

```python
pound = 0
while(pound!=-1):
  pound = int(input('파운드를 입력하세요: '))
  if(pound == -1):
    break
  print(pound,"파운드는",pound*0.453592,'킬로그램 입니다')

```
