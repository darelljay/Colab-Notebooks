### 1) if~ else 구문을 이용해서 키보드로 주민번호 뒷자리 첫 번째 수를 입력 받아 남녀를 판정하는 프로그램을 작성하시오.

힌트:주민번호 뒷자리 첫 번째 숫자가1 또는 3이면 남성이고 2 또는 4이면 여성임.

실행결과:

주민번호 뒷자리 첫 번째 숫자를 입력하세요 : 3
남성입니다!

```python
id = input('주민번호 뒷자리 첫번쨰수를 입력하세요: ')

if(id == 1 or id == 3):
  print('남성입니다!')
elif(id==2 or id==4):
  print('여성입니다!')    
```

### 2) 보통 안과에서는 시력이 0.7 이하이면 안경을 착용하는 것을 권한다고 합니다. 키보드로 시력을 입력 받아 안경 착용 여부를 판단하는 프로그램을 작성하시오.

실행결과:

당신의 시력을 입력하세요: 0.4

안경 착용을 권합니다.

```python
eyesight = float(input('당신의 시력을 입력하세요: '))

if(eyesight<=0.7):
  print('안경 착용을 권합니다.')
  
```

###3) 인터넷 카페에서 글쓰기 5개와 댓글 10개 이상을 쓴 회원에게 정회원 자격인 회원 레벨 7을 부여하고, 그렇지 않은 경우에는 준회원인 레벨 9를 부여하는 프로그램을 작성하시오.

실행결과:

글쓴 개수를 입력하세요 : 6
댓글 단 개수를 입력하세요 : 12
- 글쓴 개수 : 6, 댓글 개수 : 12
- 회원 레벨 : 7

```python
word = int(input('글쓴 새수를 입력하세요: '))
comment = int(input('댓글 개수를 입력하세요: '))

if(word >=5 and comment>=10):
  print('회원 레벨: 7')
else:
  print('회원 레벨: 6')  
```

###1) 구매 금액에 따른 할인율은 다음과 같을 때 구매 금액에 따라 지불 금액 계산하기
image.png

공식:지불금액 - 구매가 - (구매가*할인율)/100

실행결과:

물건 구매가를 입력하세요 : 8000

구매가 : 8000

지불 금액 : 7600

```python
price = int(input('물건 구매가를 입력하세요: '))

if(price<5000):
  print(price - (price*0)/100)
elif(price>=5000 or price<10000):
    print(price - (price*5)/100)
elif(price>=10000 or price<50000):
  print(price - (price*7.50)/100)  
elif(price>=50000):
  print(price - (price*10)/100)    
```

###2)점수를 입력 받아 금 점수에 해당되는 등급(A,B,C,D,E,F)을 판정하는 프로ㅡ그램을 작성하시오

90 ~ 100:A, 80 ~ 89:B, 70 ~ 79:C, 60 ~ 69:D, 0 ~ 59:F

실행결과:

점수를 입력하세요: 55

등급 : F

```python
score = int(input('점수를 입력하세요: '))

if(score>=90 or score==100):
  print('등급: A')
elif(score<=89 or score>=80):
  print('등급: B')
elif(score<=79 or score>=70):
  print('등급: C')
elif(score<=69 or score>=60):
  print('등급: D')
else:
  print('등급: F')    
```
