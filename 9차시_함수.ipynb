# 함수(for)  
파이썬의 함수는

1.일련의 코드를 담아 여러번 재사용할 수 있다.

2.인수를 매개변수로 받아 코드를 실행한 후 결과를 반환한다.  


> def 함수이름(매개변수):  
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행할 문장1  
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;수행할 문장2  
  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;return 결과값

연 습 문 제

```python
sum = 0
a = int(input("몇부터 더할까요?"))
b = int(input("몇까지 더할까요?"))
for i in range(a,b+1):
  sum+=i
print(f'{a}부터 {b}까지 더한 합계는:',sum) 
```

```python
a = int(input("몇 갱의 숫자를 더할까요"))
total = 0
for i in range(a):
  b = int(input("정수입력"))
  if b%2==0:
      total+=b
    
print(total)
```

```python
sum=0
while True:
  a = int(input("숫자: "))
  if a==0:
     break
  sum+=a

print(sum)  
```

```python
score_list =[]
for i in range(5):
  score = int(input("score"))
  score_list.append(score)
score_list.sort()
print("1등 점수: ",score_list[4])
print("2등 점수: ",score_list[3])
```

```python
#매개 변수가 여러개인 경우 (매개 변수 순서를 지켜야 함)
def introduce(name,age):
  print(f'안녕하세요 저의 이름은{name}이고 나이는{age}입니다.')
introduce("darell",15)  
introduce(age=17,name='홍길동')  
```

```python
#계산하는 함수 만들어 보기 (add(), subtract())
def add(a,b):
  print(f'덧셈:{a} + {b} = {a+b}')
def subtract(a,b):
  print(f'뺄셈:{a} - {b} = {a-b}')
  
a =int(input('첫번째 숫자: '))
b = int(input('두번째 숫자: '))
add(a,b)
subtract(a,b)
```

```python
# 2개의 정수 start, end까지의 합을 구하는 함수
def get_sum(a,b):
  sum = 0
  for i in range(a,b+1):
    sum+=i
  print(sum)
get_sum(1,10)
```
