3. 리스트를 구성하는 원소를 바꿀 수 있다.

## 2) 리스트(list) 다루기 기본

리스트(list)에서 원소의 가져오기, 수정하기, 추가하기, 제거하기

**리스트를 구성하는 원소 가져오기**

```python
a = [1,2,3,4,5]
a
```

```python
# a 리스트의 첫번째 위치의 값 얻기
element_2 = a[0]
element_2
```

```python
# 리스트의 마지막 값 얻기
element_3 = a[-1]
element_3
```

```python
#리스트의 첫번째 값와 세번째 값 더하기
a[0] + a[2]
```

```python
# 리스트의 범위를 지정하여 값 얻기 ..[2,3] 출력
elememt_3 = a[1:3] 
elememt_3
```

```python
# 리스트의 세번째 이후 값 얻기   [3,4,5]
element_4 = a[2:5]
element_4
```

```python
# 리스트의 범위에서 step으로 건너뛰며 값 얻기  [1,3,5]
element_5 =a[::2]
element_5
```

**리스트의 특정 원소 확인하기**

```python
lst = [1,2,3]
```

```python
# 리스트에 요소 추가 ( 리스트 + 리스트)   결과값 [1,2,3,4,5]
lst= lst+[4,5]
lst
```

```python
# 멤버연산자 ( 해당 값이 리스트 안에 있을경우 in  없을 경우 not in)
print(1 in(1,2,3,4))
print(5 in[1,2,3,4])
print('p' in 'python')
print('a' in 'python')
print(1 not in(1,2,3,4))
print(5 not in[1,2,3,4])
print('p' not in 'python')
'a' not in 'python'
```

**리스트의 특정 원소 수정하기**

```python
# 리스트의 특정 위치의 값을 수정하기
a = [1,2,3]
a[2] = 4
a
```

**리스트에 새로운 원소 추가하기**

```python
a = [1,2,3]
# 리스트의 가장 뒤에 값 추가하기 (append)
a.append(4)
a
```

```python
# 리스트의 가장 뒤에 리스트를 원소로 추가하기 (append)
a.append([5,6])
a
```

```python
# 리스트의 가장 뒤에 리스트로 이뤄진 값 추가하기(extend)
a = [1,2,3]
a.extend([4,5])
a
b = [6,7]
a.extend(b)
a
```

```python
# 리스트의 특정 위치에 특정 값 추가하기 insert(인덱스, 값)
a.insert(0,4)
a
a.insert(3,5)
a
```

**리스트의 특정 원소 제거하기**

```python
a = [1,2,3]
# 특정 위치의 값 제거   del 리스트명[index]
del a[1]
a
a = [1,2,3,4,5]
del a[2:]
a
```

```python
# 특정 항목값 제거    리스트명.remove(값)
a = [1,2,3,1,2,3]
a.remove(3)
a
a.remove(3)
a
```

```python
# 가장 마지막 원소을 반환하고, 해당 원소는 리스트에서 제거하기  리스트명.pop(index)
a = [1,2,3]
a.pop()
a = [1,2,3]
a.pop(1)
```

```python
#모든 요소 제거      리스트명.clear()
a.clear
```

```python
# 해당 값이 리스트 안에 없을 경우
lst = [1,2,3,4,5]

#lst.remove(6)
#del lst[5]
lst

lst = []
lst.pop()
```
