다섯 과목의 성적의 합계와 평균 구하기

```python
scores = [ 80, 92, 99, 75, 87]
sum = 0

for a in scores:
      sum+=a
avg = sum/len(scores)      
print("합계:",sum)
print('평균',avg)
```

###4)리스트에 while문 사용하기

```python
scores = [ 80, 92, 99, 75, 87]
sum = 0
i = 0
while i < len(scores):
  sum+=scores[i]
  i = i+1
avg = sum/len(scores)
print(sum,avg)    
```

###7) 연습문제

키보드로 성적을 입력 받아 리스트에 저장한 다음 평균을 구하는 프로그램을 작성하시오. 단, -1을 입력했을 때 성적 입력을 종료함.

실행결과

성적을 입력하세요(종료 시 -1 입력) : 55

성적을 입력하세요(종료 시 -1 입력) : 33

성적을 입력하세요(종료 시 -1 입력) : 33

성적을 입력하세요(종료 시 -1 입력) : 55

성적을 입력하세요(종료 시 -1 입력) : -1

입력된 성적 :  [55, 33, 33, 55]

평균 : 44.00

```python
scores = []
sum = 0
a = 0
while(a!=-1):
  a = int(input("성적을 입력하세요(종루 시 -1 입력)"))
  scores.append(a)
  if(a == -1):
    scores.remove(a)
  sum+=a
avg = sum/len(scores)    
print('입력된 성적:',scores)  
print('평균:',avg)
```
