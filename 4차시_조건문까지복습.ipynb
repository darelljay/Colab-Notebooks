# 1. 문자열 출력을 위한 포맷팅


---





> "이번 english점수는 90입니다."



```python
score = 90
subject = "english"
```

**방법 1. 문자열 출력( , 사용)**

```python
print('이번',subject,'점수는',score,'입니다')
```

**방법 2. 문자열 더하기 (비추천)**

```python
print('이번'+subject+'점수는'+str(score)+'입니다')
```

**방법 3. f-문자열 이용하기**

```python
print(f'이번{subject}점수는{score}입니다.')
```

**방법 4. . % 사용하기**

```python
print('이번 %s점수는 %d입니다'%(subject,score))
```

**방법 5. .format 함수 사용하기**

```python
print('이번 {}점수 {}입니다'.format(subject,score))
```

## 1.1 print()문의 옵션

*   \n   : 새줄을 시작한다.
*   \t   : 문자열 사이에 탭 간격을 준다.
*   \\   : 역슬래시(\)문자를 그대로 표현한다.
*   \'   : 작은따옴표(')문자를 그대로 표현한다.
*   \"   : 큰따옴표(')문자를 그대로 표현한다.


```python
print("funny\nday")
print("funny\tday")
print("\"funnyday\"")
```

###  비교 조건 연산자를 사용한 조건식

```python
score = 0
#만약 변수 x가 30보다 크거나 같으면 socore 10점 상승
x=32
if(x >= 30):
  score = score +10
```

```python
#만약 변수 j는 3의 배수이면 10점 상승
j = 6
if (j%3==0):
  score = score +10
```

```python
#만약 i가 홀수이면 10점 상승
i = 1
if(i%2==1):
  score = score + 10
```

```python
#x는 10보다 크고 20 이하이면 10점 상승
x = 15
if(x>10 and x<=20):
  score = score +10
```

```python
#i는 2의 배수 또는 3의 배수이면 10점 상승
i = 4
if(i%2==0 or i%3==0):
  score = score + 10
```

```python
#y가 0보다 작거나 100보다 크면 10점 상승
y = 120
if(y>0 or y>100):
  score = score +10
```

[문제] 학생의 점수를 입력 받아 학점을 출력하는 프로그램을 작성하세요.


> 90점 이상이면 A, 80점 이상이면 B, 0점이면 F, 나머지는 C



```python
score=input("점수를 입력하세요")
if int(score)>=90:
  print('A')
elif int(score)>=80:
    print('B')
elif int(score)<=0:
      print('F')
else:
      print("c")
```

[문제]나이가 10살 이상이고 키가 110cm 이상이면 놀이 기구를 탈 수 있다. 나이와 키를 입력 받아 알려주세요. 

```python
age = input("나이를 입력하세요")
height = input("키를 입력하세요")
if int(age)>=10 and int(height)>=110:
  print("놀이기구 탑승이 가능하십니다.")
else:
    print("놀이기구를 못타십니다")
```

[문제]4로 나누어지면서 100으로 나누어지지 않는 년도는 윤년이다. 400으로 나누어 떨어지는 년도는 윤년이다.
