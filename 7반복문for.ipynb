### 1) 1~10 정수 합계 구하기
실행결과:

1~10의 합계 : 55

```python
sum = 0
for i in range(1,11):
    sum+=i
print(sum)  
```

### 2) 500~1000의 수 중에서 3의 배수 합계 구하기
실행결과:

501

504

507

...

993

996

999

합계 : 125250

```python
sum = 0
for i in range(500,1001):
  if(i%3==0):
    sum+=i
print(sum)    
```

###3) 1~1000 의 자연수 중에서 5의 배수가 아닌 수의 합계 구하기
실행결과:

합계 : 400000

```python
sum = 0
for i in range(1,1000):
  if(i%5!=0):
    sum+=i
print('합계:',sum)    
```

###5)전화번호에서 하이픈(-) 제거하기

실행결과:

하이픈(-)을 포함하여 전화번호를 입력하세요 : 010-555-1234

0105551234

```python
num = '010-1234-5678'
result = num
for i in num:
  if '-' in num:
    result = result.replace('-','')
print(result)    
```

###6)공백 대신 밑줄(_) 삽입하기

실행결과:

문장을 입력하세요 : 안녕 친구야 만나서 반가워

안녕_친구야_만나서_반가워

```python
a = '안녕 친구야 만나서 반가워'
str = a
for i in a: 
  if ' ' in a:
    str = str.replace(' ','_')
print(str)    
```

###7) 1~10 까지 리터 갤런 온스 리스트 만들기

공식

갤런 = 리터 x 0.264
온스 = 리터 x 33.814

실행결과:

------------------------------
    리터     갤런     온스
------------------------------
    1.00     0.26    33.81
    3.00     0.79   101.44
    5.00     1.32   169.07
    7.00     1.85   236.70
    9.00     2.38   304.33
------------------------------


```python

print('리터 갤런 온스')
print('---'*100)
for i in range(1,10,2):
  lit = i * 0.264
  ons = i*33.814
  print("%d %.2f %.2f"% (i,lit,ons))
```
